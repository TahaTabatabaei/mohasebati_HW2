```python
import pandas as pd
data = pd.read_csv('train.csv')
data.head()
```

```python
data = data.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])

x = data.drop('satisfaction', axis=1)
y = data['satisfaction']    

data.head()
```

```python
from sklearn.linear_model import Perceptron
trainPer = Perceptron(random_state=0)
trainPer.fit(x.values, y.values)
```

```python
dataTest = pd.read_csv('test.csv')
dataTest.head()

dataTest = dataTest.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    dataTest[column] = encoder.fit_transform(dataTest[column])

x2 = dataTest.drop('satisfaction', axis=1)
y2 = dataTest['satisfaction']    
```

```python

from sklearn.metrics import accuracy_score
accuracy_score(y2.values, trainPer.predict(x2.values))
```

```python
from sklearn.kernel_approximation import RBFSampler
from sklearn.linear_model import SGDClassifier

rbf_feature = RBFSampler(gamma=0.0001, random_state=0)
X_features = rbf_feature.fit_transform(dataTest)

clf = SGDClassifier(max_iter=2000)
clf.fit(X_features,  y2.values)
clf.score(X_features, y2.values)
```
