```python
import pandas as pd
import numpy as np
from sklearn.pipeline import make_pipeline

trainData = pd.read_csv('train.csv')
trainData.head()
```

## Simple perceptron
(phase 1)

```python
from sklearn.preprocessing import MinMaxScaler, Normalizer
data = trainData.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])

data = data.drop('Gender', axis=1)
x = data.drop('satisfaction', axis=1)
y = data['satisfaction']    

data.head()
mm = make_pipeline(MinMaxScaler(), Normalizer())
x = mm.fit_transform(x)
```

```python
from sklearn.linear_model import Perceptron
trainPer = Perceptron(random_state=0)
trainPer.fit(x, y.values)
```

```python
testData = pd.read_csv('test.csv')
testData.head()
```

```python

dataTest = testData.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    dataTest[column] = encoder.fit_transform(dataTest[column])

dataTest = dataTest.drop('Gender', axis=1)
x2 = dataTest.drop('satisfaction', axis=1)
y2 = dataTest['satisfaction']  

dataTest.head()
mm = make_pipeline(MinMaxScaler(), Normalizer())
x2 = mm.fit_transform(x2)
```

```python
from sklearn.metrics import accuracy_score
accuracy_score(y2.values, trainPer.predict(x2))
```

## kernel trick
(phase 2)

### kerenel rbf

```python
import pandas as pd

data = trainData.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])

data = data.drop('Gender', axis=1)
x = data.drop('satisfaction', axis=1)
y = data['satisfaction']    

data.head()
mm = make_pipeline(MinMaxScaler(), Normalizer())
x = mm.fit_transform(x)
```

```python
from sklearn.kernel_approximation import RBFSampler
rbf_feature = RBFSampler(gamma=1, random_state=1)

x_new_rbf = rbf_feature.fit_transform(x)
print(x.shape)
print(type(x))
print(x_new_rbf.shape)
print(type(x_new_rbf))

# x_new_rbf = pd.DataFrame(x_new_rbf)

from sklearn.linear_model import Perceptron
trainPer = Perceptron(random_state=0)
trainPer.fit(x_new_rbf, y.values)
```

```python
testData = pd.read_csv('test.csv')
testData.head()

dataTest = testData.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    dataTest[column] = encoder.fit_transform(dataTest[column])

dataTest = dataTest.drop('Gender', axis=1)
x2 = dataTest.drop('satisfaction', axis=1)
y2 = dataTest['satisfaction']  

dataTest.head()
mm = make_pipeline(MinMaxScaler(), Normalizer())
x2 = mm.fit_transform(x2)


x2_new_rbf = rbf_feature.fit_transform(x2)
print(x2.shape)
print(type(x2))
print(x2_new_rbf.shape)
print(type(x2_new_rbf))

# x2_new_rbf = pd.DataFrame(x2_new_rbf)

from sklearn.metrics import accuracy_score
accuracy_score(y2.values, trainPer.predict(x2_new_rbf))
```

### kernel polynomial

```python
import pandas as pd

data = trainData.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    data[column] = encoder.fit_transform(data[column])

data = data.drop('Gender', axis=1)
x = data.drop('satisfaction', axis=1)
y = data['satisfaction']    

data.head()
mm = make_pipeline(MinMaxScaler(), Normalizer())
x = mm.fit_transform(x)


from sklearn.kernel_approximation import PolynomialCountSketch
from sklearn.svm import LinearSVC
from sklearn.pipeline import make_pipeline   

print(x.shape[1])
pipeline = make_pipeline(
            PolynomialCountSketch(n_components=1000, degree=3),
            LinearSVC(dual="auto"),
        )

pipeline.fit(x, y.values)




testData = pd.read_csv('test.csv')
testData.head()

dataTest = testData.drop(['Unnamed: 0', 'id'], axis=1).dropna()

from sklearn.preprocessing import LabelEncoder
for column in ['Gender', 'Customer Type', 'Type of Travel', 'Class', 'satisfaction']:
    encoder = LabelEncoder()
    dataTest[column] = encoder.fit_transform(dataTest[column])

dataTest = dataTest.drop('Gender', axis=1)
x2 = dataTest.drop('satisfaction', axis=1)
y2 = dataTest['satisfaction']  

dataTest.head()
mm = make_pipeline(MinMaxScaler(), Normalizer())
x2 = mm.fit_transform(x2)

pipeline.score(x2, y2.values)
```
